# FIRST WE WILL SCRAPE THE DATA OF 5000 USED CARS & THEN WILL MERGE THE DATA INTO ONE CSV FILE 

```python
#Importing the needed libraries

import pandas as pd
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
```

```python
#Creating an instance of the chrome driver to get to the url

driver = webdriver.Chrome(r'E:/CHROME-DRIVER/chromedriver_win32/chromedriver.exe')
```

```python
#Getting into the website

driver.get('https://www.olx.in/en-in/mumbai_g4058997/cars_c84?isSearchCall=false')
```

```python
#Creating empty lists so that we can fetch the data

Price=[]
Year=[]
Km=[]
Company=[]
Location=[]
Date=[]
```

```python
#Fetching the data from olx for Price

for p in driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li/a/div[1]/div[2]/span"):
    Price.append(p.text.replace("â‚¹","").replace(",",""))
```

```python
#Printing the price

Price
```

```python
#Converting the column into integer type

for i in range(0, len(Price)):
    Price[i] = int(Price[i])
```

```python
#Printing the Price

Price
```

```python
#Fetching data from olx for Year

for y in driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li/a/div[1]/div[2]/div[1]"):
    Year.append(y.text.split('-')[0])
```

```python
#Converting the column into integer type

for i in range(0, len(Year)):
    Year[i] = int(Year[i])
```

```python
#Converting year into integer type

Year
```

```python
#Fetching data from olx for Km used, Company, Location, Date of posting

for k in driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li/a/div[1]/div[2]/div[1]"):
    Km.append(k.text.split('-')[1])
for c in driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li/a/div[1]/div[2]/div[2]"):
    Company.append(c.text)
for l in driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li/a/div[1]/div[2]/div[3]"):
    Location.append(l.text)
for d in driver.find_elements_by_xpath("//*[@id='container']/main/div/div/section/div/div/div[5]/div[2]/div/div[2]/ul/li/a/div[1]/div[2]/div[3]/span"):
    Date.append(d.text)
```

```python
#Making a dataframe named Cars storing 500 data into that

CARS=pd.DataFrame({})
CARS['PRICE']=Price[0:500]
CARS['YEAR']=Year[0:500]
CARS['KM']=Km[0:500]
CARS['COMPANY']=Company[0:500]
CARS['LOCATION']=Location[0:500]
CARS['POSTED-DATE']=Date[0:500]
```

```python
#Printing the dataframe of CARS that we have created 

CARS
```

```python
#Printing the information of the dataset
#Here we can see 2 columns are of integer type & others are of object datatype

CARS.info()
```

```python
#Closing the driver

driver.close()
```
